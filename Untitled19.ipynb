```
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```
URL = 'https://www.marketsandmarkets.com/telecom-and-IT-market-research-113.html'
page = requests.get(URL)
soup = BeautifulSoup(page.content,'html.parser')
print(soup)
```

```
results = soup.find(id='publishedreports')
print(results)
```

```
block = results.find_all('tr',class_='alt')
print(block)
```

```
block_1 = results.find_all('tr')
print(block_1)
```

```
print(block_1[1])
```

```
titles=[]
dates=[]
links=[]
dess =[]
USDs=[]
grows=[]
for blocks in block_1[1:]:
  bl_ck = blocks.find_all('td')
  #print(bl_ck[0])
  title = blocks.find('a')
  #print(title)
  init = str(title).find('"')
  #print(init)
  last = str(title).find('"',init+1)
  #print(last)
  link = str(title)[init+1:last]
  para = blocks.find("p")
  dot = str(para.text).find("period")
  #print(dot)
  des = str(para.text)[:dot+len("period")+1]
  #print("des=",des)
  per1 = str(para.text).find(" of ")
  per2 = str(para.text).find("%")
  grow = para.text[per1+3:per2+1]
  if len(grow)>10:
    grow = para.text[per2-4:per2+1]
  usd1 = str(para.text).find("USD")
  usd2 = str(para.text).find("billion")
  usd3 = str(para.text).find("million")
  if (usd3==-1):
    USD = para.text[usd1:usd2+7]
  else:
    USD = para.text[usd1:usd3+7]
  if link =="/Market-Reports/e-discovery-market-11881863.html":
    des = "The major players in the eDiscovery market are Microsoft(US), IBM(US), OpenText(Canada), Deloitte(US), Micro Focus(UK), ZyLAB(The Netherlands), Relativity(US), Nuix(Australia), Logikcull(US), KLDiscovery(US), Exterro(US), Thomson Reuters(Canada), Knovos(US), Casepoint(US), Nextpoint(US), DISCO(US), Veritas(US), One Discovery(US), Congruity360(US), Onna(US), Everlaw(US), Texifter(US), Allegory(US), Evichat(Canada), and Reductech(UK)."

  date = bl_ck[1]
  print(title.text)
  titles.append(title.text)
  print("date=",date.text)
  dates.append(date.text)
  print("link=",link)
  links.append(link)
  print("para=",des,"\n")
  dess.append(des)
  print("grow=",grow)
  grows.append(grow)
  print("USD=",USD)
  print(usd1)
  USDs.append(USD)
```

```
print(titles)
```

```
with open('submission.csv','w',newline='') as file:
  writer = csv.writer(file)
  writer.writerow(['report_title', 'date_published', 'link', 'report_description', 'USD _value', '%_val' ])
  print(len(titles))
  for i in range(0,len(titles)):
    if(dess[i]=="The global customer journey analytics market size to grow from 8.3 billion in 2020 to USD 25.1 billion by 2026, at a Compound Annual Growth Rate (CAGR) of 20.3% during the forecast period."):
      USDs[i]= 'USD 8.3 billion'
    if(dess[i]=="The Partner Relationship Management Market is expected to grow from USD Old to USD 1,997 million by 2026, at a CAGR of 13.8% during the forecast period."):
      USDs[i]= 'USD 1,997 million'
    writer.writerow([i,titles[i],dates[i],"https://www.marketsandmarkets.com"+links[i],dess[i],USDs[i],grows[i]])
  print(file)
```
